```python
x= data.iloc[:,:-1].values
```

```python
print(x)
```

# Using the elbow method to find the optimal number of clusters

```python
from sklearn.cluster import KMeans
wcss = []
for i in range(1,11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++',
                    random_state = 0)
    kmeans.fit(x)
    wcss.append(kmeans.inertia_)
plt.plot(range(1, 11), wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

## We observe that the optimal number of clusters is 3

# Fitting K-Means to the dataset

```python
kmeans = KMeans(n_clusters = 3, init = 'k-means++', 
                random_state = 0)
y_kmeans = kmeans.fit_predict(x)
```

```python
y_kmeans
```

# Cross-Tabulation

```python
ct=pd.DataFrame({'Species':data['Species'],'kmeans':y_kmeans})
ct
```

```python
ct['kmeans'].value_counts()
```

```python
crosstab=pd.crosstab(ct['kmeans'],ct['Species'])
sns.heatmap(crosstab, annot=True)
```

# For the above heatmap we assign cluster 0 to Iris-virginica, cluster 1 to Iris-setosa and cluster 2 to Iris-versicolor.

# Here, 16 out of 150 observation have been wrongly classified.

# Visualising the clusters, using features are 'Petal length' and 'Petal width'

```python
# Visualising the clusters - On the first two columns
plt.scatter(x[y_kmeans == 0, 2], x[y_kmeans == 0, 3], 
            s = 80, c = 'red', label = 'Iris-virginica')
plt.scatter(x[y_kmeans == 1, 2], x[y_kmeans == 1, 3], 
            s = 80, c = 'lime', label = 'Iris-setosa')
plt.scatter(x[y_kmeans == 2, 2], x[y_kmeans == 2, 3],
            s = 80, c = 'hotpink', label = 'Iris-versicolor')
plt.scatter(kmeans.cluster_centers_[:, 2], kmeans.cluster_centers_[:, 3],marker='*',s = 130, c = 'blue', label = 'Centroids')
plt.title("Iris Clusters")
plt.xlabel("Petal Length(cm)")
plt.ylabel("Petal Width(cm)")
plt.legend()
plt.show()
```
